```python
import requests

host = "http://api.coincap.io"

url_doge = f"{host}/v2/assets/dogecoin/history?interval=m5"
url_btc = f"{host}/v2/assets/bitcoin/history?interval=m5"
url_eth = f"{host}/v2/assets/ethereum/history?interval=m5"

payload={}
headers = {}

response_doge = requests.request("GET", url_doge, headers=headers, data=payload)
response_eth = requests.request("GET", url_eth, headers=headers, data=payload)
response_btc = requests.request("GET", url_btc, headers=headers, data=payload)

print(response_doge.text)
```

```python
def custom_predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name='Predictions')
    combined = pd.concat([test['Target'], pd.Series(preds, index=test.index)], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=400, step=50):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = custom_predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(doge, model, predictors)
```

```python
from decimal import Decimal
horizons = [2, 5, 10, 25, 50, 100, 250]
new_predictors = []

for horizon in horizons:
    for df in [doge, btc, eth]:
        name = 'doge' if df is doge else 'btc' if df is btc else 'eth'
        rolling_averages = df['priceUsd'].rolling(horizon).mean()

        ratio_column = f"{name}_Price_Rate_{horizon}"
        df[ratio_column] = df['priceUsd'] / rolling_averages

        trend_column = f"{name}_Trend_{horizon}"
        df[trend_column] = df.shift(1).rolling(horizon).sum()['Target']

        new_predictors += [ratio_column, trend_column]
```

```python
window_size = 100  # Number of data points in each window

train_windows = []
test_windows = []

for i in range(len(merged_df) - window_size):
    train_window = merged_df.iloc[i:i+window_size]
    test_window = merged_df.iloc[i+window_size]
    
    train_windows.append(train_window)
    test_windows.append(test_window)

# Example usage:
train = pd.concat(train_windows)
test = pd.concat(test_windows)
```

```python
def custom_predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    probs = model.predict_proba(test[predictors])[:, 1]
    preds = model.predict(test[predictors])
    preds[probs >= 0.6] = 1
    preds[probs < 0.6] = 0
    preds = pd.Series(preds, index=test.index, name='Predictions')
    probabilities = pd.Series(probs, index=test.index, name='Probability')
    combined = pd.concat([test['Target'], preds, probabilities], axis=1)
    return combined
```
